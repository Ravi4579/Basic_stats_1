# **Standardization of Numerical Variables**

```python
'''
Concept of standardization (z-score normalization).

1)z-score makes the use of the mean and standard deviation of the dataset in order to specify the relative location of the measurement.
2)It represents the distance between the given data point and the mean,expressed in standard deviation
  This score is known as "standardization" of the data point
3)The large Z-scores tell us that the measurement is larger than almost all measurement in the dataset.
4)Simillary,a small Z-score tells us that the measurment is small than all other measurments.
5)If a score is "0",then the observation lies on the mean

The formula used is:
Z=x-𝜇/𝜎

where:
𝑥=x is the value of the feature,
𝜇=μ is the mean of the feature,
𝜎=σ is the standard deviation of the feature.

'''
```

```python
#columns only with numerical values
df.select_dtypes(include=['number']).columns
```

```python
#dataset with numeric values
dfn = df.select_dtypes(include=['number'])
```

```python
dfn
```

```python
#	Standardize the numerical columns using the formula: z=x-mu/sigma
dfns = dfn.apply(lambda x: (x - x.mean()) / x.std())
```

```python
dfns
```

Show before and after comparisons of the data distributions.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Used for loop for showing distribution
#before and after comparisons of the data distributions.

for column in dfns:
#Before Standardization
  fig, axes = plt.subplots(1, 2, figsize=(15, 4))
  sns.distplot(df[column], ax=axes[0])
  axes[0].set_xlabel(column)
  axes[0].set_ylabel('Count')
  axes[0].set_title('Data distributon for '  + column +',Before Standardization')

#After Standardization
  sns.distplot(dfns[column], ax=axes[1])
  axes[1].set_xlabel(column)
  axes[1].set_ylabel('Count')
  axes[1].set_title('Data distributon for ' + column + ', After Standardization')

plt.show()
```

```python
dfc
```

```python
df_dummies.head()
```

```python
#converting into int type
df_dummies=df_dummies.astype(int)
```
